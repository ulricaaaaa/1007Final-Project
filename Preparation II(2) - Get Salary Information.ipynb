# Preparation II(2) - Get Salary Information

In this part we extract mainly salary data of different companies using Glassdoor API.

Here we are also using AidaSehic's list. Four lists on different categories are used to get more data.

```python
import  requests
import urllib.request
import csv
import pandas as pd 
import numpy as np
import json
from bs4 import BeautifulSoup
import time
```

```python
employer_df = pd.read_csv("Employer.csv")
jobTitle_df = pd.read_csv("jobTitle.csv")
designation_df=pd.read_csv("Designation.csv")
designation_company_df=pd.read_csv('designation_company.csv')
print(employer_df.shape[0])
print(jobTitle_df.shape[0])
print(designation_df.shape[0])
print(designation_company_df.shape[0])
```

## The 1st list

```python
URLs=[]
for i in range(employer_df.shape[0]):
    parametars={'t.p':25738,'t.k':'iRCtcWJQamE','userip':'0.0.0.0','format':'json','v':1,'action':'salaries','employer':employer_df['Employer'][i]}
    r = requests.get("http://api.glassdoor.com/api/api.htm", params=parametars)
    #print(r.url)
    URLs.append(r.url)
```

```python
file = open('URLs_salaries','w');
file.write(str(URLs));
file.close();
```

```python
hdr = {'User-Agent': 'Mozilla/5.0'}
#salaries_list_nested = []
for i in range(len(URLs)-1512):
    print(i+1512)
    url = URLs[i+1512]
    req = urllib.request.Request(url,headers=hdr)
    response = urllib.request.urlopen(req)
    time.sleep(0.5)
    soup = BeautifulSoup(response, 'lxml')
    salaries_data=json.loads(soup.p.get_text())
    salaries_list_nested.append(pd.DataFrame(salaries_data['response']['employerSalaryGroup']))
```

```python
salaries_list_nested[200]
```

```python
salaries_df_nested = pd.DataFrame()
for i in range(len(salaries_list_nested)):
    salaries_df_nested=salaries_df_nested.append(salaries_list_nested[i])
salaries_df_nested
```

```python
split_salary = salaries_df_nested['salary'].apply(pd.Series) # split the salary column from salaries_df_nested
split_salary = split_salary.reset_index(drop=True) # reset index

#longName	name	salary	squareLogo	totalJobTitlesCount	totalSalariesCount
id_salary = [] # declare a null list
company_name = []

company_longName = []
companyLogo = [] 

totalJobTitlesCount = []
totalSalariesCount = [] 

salaries_df_1 = pd.DataFrame()

for i in list(range(len(split_salary))):
        salaries_df_1 = salaries_df_1.append(split_salary.iloc[i,:].apply(pd.Series))
        id_salary.append(np.repeat(salaries_df_nested.iloc[i]['id'], len(split_salary.columns))) 
        company_name.append(np.repeat(salaries_df_nested.iloc[i]['name'], len(split_salary.columns)))
        company_longName.append(np.repeat(salaries_df_nested.iloc[i]['longName'], len(split_salary.columns)))
        companyLogo.append(np.repeat(salaries_df_nested.iloc[i]['squareLogo'], len(split_salary.columns)))
        totalJobTitlesCount.append(np.repeat(salaries_df_nested.iloc[i]['totalJobTitlesCount'], len(split_salary.columns)))
        totalSalariesCount.append(np.repeat(salaries_df_nested.iloc[i]['totalSalariesCount'], len(split_salary.columns)))
        

a = [j for i in id_salary for j in i]
salaries_df_1['id'] = a

b = [j for i in company_name for j in i]
salaries_df_1['company_name'] = b

c = [j for i in company_longName for j in i]
salaries_df_1['company_longName'] = c

d = [j for i in companyLogo for j in i]
salaries_df_1['companyLogo'] = d

e = [j for i in totalJobTitlesCount for j in i]
salaries_df_1['totalJobTitlesCount'] = e

f = [j for i in totalSalariesCount for j in i]
salaries_df_1['totalSalariesCount'] = f


salaries_df_1 = salaries_df_1.dropna(subset=['jobTitle']) # dropping NA
salaries_df_1=salaries_df_1.reset_index(drop=True) # Reset the index
salaries_df_1 = salaries_df_1.drop(0, axis=1)
```

```python
salaries_df_1.to_csv('salaries1_companies.csv', sep = ',', index = False) # create csv for salaries
salaries_df_1
```
